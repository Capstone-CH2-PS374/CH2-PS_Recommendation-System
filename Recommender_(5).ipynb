```python
import pandas as pd
import tensorflow as tf
import  tensorflow_recommenders as tfrs
import numpy as np
from typing import Dict, Text
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
grouping = relawan_df['Volunteer Name']
names = []
for name in grouping:
  names.append(name.replace(" ", ""))
print(names)
```

```python
tokenizer = Tokenizer(num_words= len(max(names)))
tokenizer.fit_on_texts(names)
print(tokenizer.word_index)
```

```python
ids = []
for item in relawan_df["Volunteer Name"]:
  for key in tokenizer.word_index:
    if item.replace(" ", "").lower() == key:
      ids.append(tokenizer.word_index[key])
event_df
```

```python
event_cp = event_df.copy()
event_ids = []
for item in relawan_df["Skills"]:
    found = False
    for item_rel in event_cp["Qualifications"]:
        if item == item_rel:
            found = True
            break
    if found is False:
        last_id = event_df.iloc[-1, 0]
        last_id += 1
        event_df = pd.concat([event_df, pd.DataFrame([{
            "Event_id": last_id,
            "Category": None,
            "Location": None,
            "Qualifications": item,
            "Age": None,
        }])])
```

```python
for item in relawan_df["Skills"]:
    for index, item_rel in enumerate(event_df["Qualifications"]):
        if item == item_rel :
            event_ids.append(event_df.iloc[index, 0])
            break
```

```python
relawan_df["event_id"] = event_ids
```

```python
relawan_df["ID"] = ids
```

```python
relawan_df['ID'] = relawan_df['ID'].astype(str)
relawan_df['event_id'] = relawan_df['event_id'].astype(str)
event_df['Event_id'] = event_df['Event_id'].astype(str)
```

```python
relawan = tf.data.Dataset.from_tensor_slices(relawan_df[['ID', 'event_id']].values)
qualifications = tf.data.Dataset.from_tensor_slices(event_df['Event_id'])
```

```python
relawans = relawan.map(lambda x: {
    "id": x[0],
    "event_id": x[1]
})

qualifications = qualifications.map(lambda x:
    x
)
```

```python
tf.random.set_seed(42)
shuffled = relawans.shuffle(2_000, seed=42, reshuffle_each_iteration=False)

train = shuffled.take(2_500)
test = shuffled.skip(2_500)
```

```python
qual = qualifications.batch(32)
ids = relawans.batch(32).map(lambda x: x["id"])

unique_qual = np.unique(np.concatenate(list(qual)))
unique_relawans = np.unique(np.concatenate(list(ids)))

unique_qual[:10]
```

```python
embedding_dimension = 32
```

```python
relawans_model = tf.keras.Sequential([
  tf.keras.layers.StringLookup(
      vocabulary=unique_relawans, mask_token=None),
  # We add an additional embedding to account for unknown tokens.
  tf.keras.layers.Embedding(len(unique_relawans) + 1, embedding_dimension)
])
```

```python
qual_model = tf.keras.Sequential([
  tf.keras.layers.StringLookup(
      vocabulary=unique_qual, mask_token=None),
  tf.keras.layers.Embedding(len(unique_qual) + 1, embedding_dimension)
])
```

```python
metrics = tfrs.metrics.FactorizedTopK(
  candidates=qualifications.batch(32).map(qual_model)
)
```

```python
loss = tfrs.tasks.Retrieval(
  metrics=metrics
)
```

```python
class RecipeModel(tfrs.Model):

  def __init__(self, relawans_model, qual_model):
    super().__init__()
    self.qual_model: tf.keras.Model = qual_model
    self.relawans_model: tf.keras.Model = relawans_model
    self.task: tf.keras.layers.Layer = loss

  def compute_loss(self, features: Dict[Text, tf.Tensor], training=False) -> tf.Tensor:
    # We pick out the user features and pass them into the user model.
    relawans_embedding = self.relawans_model(features["id"])
    # And pick out the movie features and pass them into the movie model,
    # getting embeddings back.
    positive_qual_embeddings = self.qual_model(features["event_id"])

    # The task computes the loss and the metrics.
    return self.task(relawans_embedding, positive_qual_embeddings)
```

```python
model = RecipeModel(relawans_model, qual_model)
model.compile(optimizer=tf.keras.optimizers.Adagrad(learning_rate=0.1))
```

```python
cached_train = train.shuffle(2_000).batch(32).cache()
cached_test = test.batch(32).cache()
```

```python
model.fit(cached_train, epochs=20)
```

```python
model.evaluate(cached_test, return_dict=True)
```

```python
# Create a model that takes in raw query features, and
index = tfrs.layers.factorized_top_k.BruteForce(model.relawans_model)
# recommends movies out of the entire movies dataset.
index.index_from_dataset(
  tf.data.Dataset.zip((qualifications.batch(32), qualifications.batch(32).map(model.qual_model)))
)

# Get recommendations.
_, titles = index(tf.constant(["10"]))
print(f"Recommendations for user 10: {titles[0, :100]}")
```
